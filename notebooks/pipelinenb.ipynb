```python
res = requests.post(biencoder, json=samples)
```

```python
res.ok
```

```python
res = res.json()
```

```python
len(res['encodings'])
```

```python
def vector_decode(s, dtype=np.float32):
    buffer = base64.b64decode(s)
    v = np.frombuffer(buffer, dtype=dtype)
    return v
```

```python
[vector_decode(e) for e in res['encodings']]
```

```python
indexer = 'http://localhost:30301/api/indexer/search'
```

```python
body = {
    'encodings': res['encodings'],
    'top_k': 10
}
```

```python
res_indexer = requests.post(indexer, json=body)
```

```python
res_indexer.ok
```

```python
res_indexer.json()
```

```python
candidates = res_indexer.json()
candidates
```

```python
[len(c) for c in candidates]
```

```python
crossencoder = 'http://localhost:30302/api/blink/crossencoder'
```

```python
body = {
    'samples': samples,
    'candidates': candidates
}
```

```python
res = requests.post(crossencoder, json=body)
```

```python
res.ok
```

```python
cross_candidates = res.json()
cross_candidates
```

```python
topent = []
for c, sam in zip(cross_candidates, samples):
    topent.append({
        'max_bi': c[0]['bi_score'],
        'max_cross': c[0]['score'],
        'mention': sam['mention'],
        'title': c[0]['title']
    })
```
